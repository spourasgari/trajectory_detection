```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_dir = '/home/sina/Desktop/SharedFolder/NewTest/Checked'

i = 1

# datapoints_no = {}

for root, dirs, files in os.walk(base_dir):
    for file in files:
        print('Working on the file:', file)

        file_path = os.path.join(root, file)

        print('Path:', file_path)

        os.system(f"python pose_detect.py --video_path {file_path}")

        file_name = file.split(".")[0]
        output_dir = f"./Recorded Datasets/{file_name}"

        waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_full_raw.csv"), delimiter=",", comments=None, dtype=str, skiprows=1)
        waypoints_10 = np.loadtxt(os.path.join(output_dir, "pose_waypoints_sampled_10hz_raw.csv"), delimiter=",", comments=None, dtype=str, skiprows=1)
        waypoints = np.array([row for row in waypoints if "None" not in row], dtype=float)
        waypoints_10 = np.array([row for row in waypoints_10 if "None" not in row], dtype=float)

        datapoints_no[file_name] = len(waypoints)

        # Plot waypoints
        plt.plot(waypoints_10[:, 2], waypoints_10[:, 3], marker="o", linestyle="-", color="blue")
        plt.title(f"Person's Path: {file_name}")
        plt.xlabel("X Position (cm)")
        plt.ylabel("Y Position (cm)")
        plt.xlim(0, 433)
        plt.ylim(0, 391)
        plt.gca().invert_yaxis()  # Invert Y to match video coordinates
        plt.gca().invert_xaxis()  # Invert X to match video coordinates
        plt.grid(True)
        plt.show()

        i += 1

        # if i >= 10:
        #     break
```

```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

datapoints_pd = pd.DataFrame.from_dict(datapoints_no, orient='index', columns=['No of Datapoints'])
datapoints_pd.sort_index(inplace=True)
```

```python
grouped = datapoints_pd.groupby(datapoints_pd.index.str[:6]).sum()

# Display the result
print(grouped)
```

```python
datapoints_pd.to_csv('datapoints.csv', index=True)
```
