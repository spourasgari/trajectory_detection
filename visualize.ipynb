```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Load waypoints
file_name = "inlab1"
output_dir = f"./{file_name}"

# waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints.txt"), delimiter="\t") # Includes None values
waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_sampled.txt"), delimiter="\t")
# waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_smoothed.txt"), delimiter="\t")

# Plot waypoints
plt.plot(waypoints[:, 2], waypoints[:, 3], marker="o", linestyle="-", color="blue")
plt.title("Person's Foot Movement Path")
plt.xlabel("X Position (cm)")
plt.ylabel("Y Position (cm)")
plt.xlim(0, 350)
plt.ylim(125, 375)
plt.gca().invert_yaxis()  # Invert Y to match video coordinates
# plt.gca().yaxis.tick_right()  # Move Y-axis ticks to the right
plt.gca().invert_xaxis()  # Invert X to match video coordinates
# plt.gca().xaxis.tick_top()  # Move X-axis ticks to the right
plt.grid(True)
plt.show()
```
