### Visualize InLab Videos
___*inlab3.mp4__ is not valid without a new transformation, as the video is recorded with different frame._

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# "inlab1"
# Load waypoints
file_name = "inlab1"
output_dir = f"./Recorded Datasets/{file_name}"

# waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_full.txt"), delimiter="\t", comments=None, dtype=str) # Includes None values
waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_sampled_10hz.txt"), delimiter="\t", comments=None, dtype=str)
waypoints = np.array([row for row in waypoints if "None" not in row], dtype=float)

# Plot waypoints
plt.plot(waypoints[:, 2], waypoints[:, 3], marker="o", linestyle="-", color="blue")
plt.title("Person's Foot Movement Path")
plt.xlabel("X Position (cm)")
plt.ylabel("Y Position (cm)")
plt.xlim(0, 433)
plt.ylim(0, 391)
plt.gca().invert_yaxis()  # Invert Y to match video coordinates
plt.gca().invert_xaxis()  # Invert X to match video coordinates
plt.grid(True)
plt.show()
```

### Evaluate InLab Videos

#### inlab_eval_1

```python
# Load waypoints
file_name = "inlab_eval_1"
output_dir = f"./Recorded Datasets/{file_name}"

# waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_full.txt"), delimiter="\t", comments=None, dtype=str) # Includes None values
waypoints = np.loadtxt(os.path.join(output_dir, "pose_waypoints_sampled_10hz.txt"), delimiter="\t", comments=None, dtype=str)
waypoints = np.array([row for row in waypoints if "None" not in row], dtype=float)

# Plot waypoints
plt.plot(waypoints[:, 2], waypoints[:, 3], marker="o", linestyle="-", color="blue")
plt.title("Person's Foot Movement Path")
plt.xlabel("X Position (cm)")
plt.ylabel("Y Position (cm)")
plt.xlim(0, 433)
plt.ylim(0, 391)
plt.gca().invert_yaxis()  # Invert Y to match video coordinates
plt.gca().invert_xaxis()  # Invert X to match video coordinates

P_o = np.array([193, 91, 0])
P_d = np.array([193, 391, 0])
plt.plot(*zip(P_o[:2], P_d[:2]), marker="o", linestyle="--", color="red")
plt.grid(True)
plt.show()


# Calculate deviation from the actual line
dist = []
for point in waypoints[:, 2:]:
    point = np.append(point, 0)
    dist.append(np.linalg.norm(np.cross(P_o-P_d, P_d-point))/np.linalg.norm(P_o-P_d))

print(f"The average deviation from the atual line is:\t {np.average(dist):.2f} cm")
```
